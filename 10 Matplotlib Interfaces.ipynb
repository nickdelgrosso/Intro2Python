```python
import numpy as np
import matplotlib.pyplot as plt
```

# Basic Plots in Matplotlib

Matplotlib was made originally to be very similar to the plotting functions in Matlab.  I've put some examples below to get a feel for some common operations and to get started; a great resource can be found at the Python Graph Gallery: https://www.python-graph-gallery.com/matplotlib/

```python
x = np.random.normal(size=100)
plt.plot(x);
```

## Setting the Components of a Matplotlib Figure: Stateful Interface
Matplotlib has two distinct ways of working with it:
  - **The Stateful Interface**: A "Matlab-like" way of giving commands the the plotting backend and letting it keep track of what the current figure is.
  - **The Object-Oriented Interface**: A more Pythonic way of sending commands directly to the objects that hold the object's attributes.

#### Stateful Interface Example: Line Chart

```python
plt.figure(figsize=(7, 5))
plt.subplot(1, 2, 1) # plt.subplot(no of rows, no of columns, active subplot)
plt.plot(x)
plt.title('CDF, More bins, More Labels')
plt.xlabel("X axis")
plt.ylabel('Y Axis')
plt.xlim(-3.5, 20);

plt.subplot(1, 2, 2)
plt.plot(x, label='My Data')
plt.plot(x*2, label='My Data*2')
plt.legend();
```

**Aside: plt.figure() doesn't need to be called in notebooks**

Matplotlib will automatically make a new figure for each cell.

```python
plt.plot(x);
```

### Example: Object-Oriented Interface Example: Line Chart

```python
fig = plt.figure(figsize=(7, 5))

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(x)
ax1.set_title('CDF, More bins, More Labels')
ax1.set_xlabel("X axis")
ax1.set_ylabel('Y Axis')
ax1.set_xlim(-3.5, 20);

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(x, label='My Data');
ax2.legend()
```

At first glance, this may seem just more verbose, but it has several advantages:

**1. Order-agnostic commands**

```python
fig = plt.figure(figsize=(7, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.plot(x)
ax1.set_xlim(-3.5, 20);

ax2.plot(x);

ax1.set_title('CDF, More bins, More Labels')
ax1.set_xlabel("X axis")
ax1.set_ylabel('Y Axis')
```

**2. Many options for more-convenient handling convenience methods**

```python
fig, (ax1, ax2) = plt.subplots(figsize=(7, 5), ncols=2)

ax1.plot(x)
ax1.set(
    title='CDF, More bins, More Labels',
    xlabel="X axis",
    ylabel='Y Axis',
    xlim=[-3.5, 20],
)

ax2.plot(x);
```

** 3. Reference the same objects across multiple cells **

```python
fig
```

** 4. Write functions that explicitly modify plots **

```python
def label_xaxis(ax, label):
    ax.set_xlabel(label)

ax = plt.subplot()
ax.plot(x)
label_xaxis(ax=ax, label='Hiya!')
```

### Moving from Stateful to Object-Oriented: Getting the current figure / axis

You can always call `plt.gcf()` or `plt.gca()` to get the current figure/axis as an object, providing a convenient way to go from one approach to the next.

*Note*: when doing ths, make sure it's called in the same cell as the figure you want to assign to, else you'll get a completely new figure.

```python
fig = plt.gcf()
fig
```

```python
ax = plt.gca()
ax;
```

## The plot() Function is flexible

  - **Line Chart**: a single 1D or 2D sequence (columns are plotted)
  - **Connected-Line Chart**: two 1D sequences, a 1D (time) and a 2D sequence (columns)
  - **Scatterplot**: same as connected-line chart, with the lines removed.

**Tip**: Check out the help text for plt.plot().  It is remarkably comprehensive!

```python
x = [3, 5, 2, 7, 8, 3, 1, 9]
y = [2, 6, 4, 7, 9, 6, 2, 1]

plt.figure(figsize=(15, 5))
plt.subplot(1, 4, 1)
plt.plot(x)
plt.plot(y)

z = np.array([x, y]).T
plt.subplot(1, 4, 2)
plt.plot(z)

plt.subplot(1, 4, 3)
plt.plot(x, y)

plt.subplot(1, 4, 4)
plt.plot(x, y, 'o');
```
